```python
def create_table(grouped_numbers, columns_in_order, column_name, stat):
    output = {}
    for column in columns_in_order:
        output[column] = []
    
    for i, row in grouped_numbers.iterrows():
        output[row[column_name]].append(row[stat])
    
    return pd.DataFrame.from_dict(output)
```

```python
ent_template_table = create_table(grouped_template_merged_operator_ent_data, perturbation_order, "perturbation", "avg_stat")
contr_template_table = create_table(grouped_template_merged_operator_contr_data, perturbation_order, "perturbation", "avg_stat")
```
